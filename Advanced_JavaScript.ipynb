**Problem 1**

Create a Map in JavaScript and perform the following operations:
- Add key-value pairs to the Map.
- Check if a specific key exists.
- Retrieve the value associated with a given key.
- Iterate through all key-value pairs.

```python
// Create a Map
const studentMap = new Map();

// 1. Add key-value pairs to the Map
studentMap.set("name", "Rahul");
studentMap.set("age", 21);
studentMap.set("course", "B.Tech");
studentMap.set("city", "Delhi");

console.log("Map after adding values:", studentMap);

// 2. Check if a specific key exists
const hasAge = studentMap.has("age");
console.log("Does key 'age' exist?", hasAge);

// 3. Retrieve the value associated with a given key
const studentName = studentMap.get("name");
console.log("Value for key 'name':", studentName);

// 4. Iterate through all key-value pairs
console.log("Iterating through Map:");
for (let [key, value] of studentMap) {
	console.log(`${key}: ${value}`);
}
```

**Output:**

Map after adding values: Map(4) {

  'name' => 'Rahul',

  'age' => 21,

  'course' => 'B.Tech',

  'city' => 'Delhi'

}

Does key 'age' exist? true

Value for key 'name': Rahul

Iterating through Map:

name: Rahul

age: 21

course: B.Tech

city: Delhi

**Problem 2**

Create a Map to store contact information (name, age, email, location) and implement a function to retrieve contact details by name.

```python
// Create a Map to store contacts
const contacts = new Map();

// Add contact information
contacts.set("Rahul", {
  name: "Rahul",
  age: 22,
  email: "rahul@gmail.com",
  location: "Delhi"
});

contacts.set("Priya", {
  name: "Priya",
  age: 24,
  email: "priya@gmail.com",
  location: "Mumbai"
});

contacts.set("Amit", {
  name: "Amit",
  age: 26,
  email: "amit@gmail.com",
  location: "Pune"
});

// Function to retrieve contact details by name
function getContactByName(name) {
  if (contacts.has(name)) {
    return contacts.get(name);
  } else {
    return "Contact not found!";
  }
}

// Example usage
console.log(getContactByName("Priya"));
console.log(getContactByName("John"));
```

**Output:**

{ name: 'Priya', age: 24, email: 'priya@gmail.com', location: 'Mumbai' }

"Contact not found!"


**Problem 3**

You are given an array of numbers named myArray. Create a function that takes any number of arguments and adds them to the existing array. Use the spread and rest operator.

```python
// Given array
let myArray = [1, 2, 3, 4];

// Function that takes any number of arguments and adds them to myArray
function addToArray(...numbers) {   // rest operator
  myArray = [...myArray, ...numbers];  // spread operator
}

// Example usage
addToArray(5, 6, 7);
addToArray(8, 9);

console.log(myArray);
```

**Output:**

[1, 2, 3, 4, 5, 6, 7, 8, 9]

**Problem 5**

Create two objects person1 and person2 with properties name and age. Create a function “introduce” that prints "Hello, I'm [name], and I'm [age] years old." Use the call method to make person2 introduce itself using the introduce function.

```python
// Create two objects
const person1 = {
  name: "Rju",
  age: 18
};

const person2 = {
  name: "Om",
  age: 22
};

// Introduce function (not inside any object)
function introduce() {
  console.log(`Hello, I'm ${this.name}, and I'm ${this.age} years old.`);
}

// Use call() to make person2 introduce itself
introduce.call(person2);
```

**Output:**

Hello, I'm Om, and I'm 22 years old.
